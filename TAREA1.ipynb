```python
# # Desempleo

import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Graficar la tasa de desempleo desestacionalizada
plt.plot(unemployment_df_filtered.index, unemployment_df_filtered['Unemployment_Rate_Seasonally_Adjusted'], label='Desestacionalizada')

# Graficar la tendencia-ciclo
plt.plot(unemployment_df_filtered.index, unemployment_df_filtered['Unemployment_Rate_Trend_Cycle'], marker='x', label='Tendencia-Ciclo')

plt.title('Tasa de Desempleo en México (2006 - 2024)', fontsize=16)
plt.xlabel('Fecha', fontsize=14)
plt.ylabel('Tasa de Desempleo (%)', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()

```

### INFLACIÓN

```python
import pandas as pd

# Cargar el dataset de inflación
inflation_df = pd.read_csv('cleaned_inflation_data (1).csv', index_col='Date', parse_dates=True)

# Verifica las primeras filas para asegurarte de que los datos estén cargados correctamente
inflation_df
```

```python
import matplotlib.pyplot as plt

# Configurar el tamaño de la gráfica
plt.figure(figsize=(10, 6))

# Graficar la tasa de inflación
plt.plot(inflation_df.index, inflation_df['Tasa de Inflación'], color='blue')

# Configurar los títulos y etiquetas
plt.title('Tasa de Inflación en México (1996 - 2024)', fontsize=16)
plt.xlabel('Fecha', fontsize=14)
plt.ylabel('Tasa de Inflación (%)', fontsize=14)

# Mostrar la gráfica
plt.grid(True)
plt.show()

```
