### INFLACIÓN (ACT EN CLASE)

```python
import pandas as pd

# Cargar el dataset de inflación
inflation_df = pd.read_csv('cleaned_inflation_data (1).csv', index_col='Date', parse_dates=True)

# Verifica las primeras filas para asegurarte de que los datos estén cargados correctamente
inflation_df
```

```python
import matplotlib.pyplot as plt

# Configurar el tamaño de la gráfica
plt.figure(figsize=(10, 6))

# Graficar la tasa de inflación
plt.plot(inflation_df.index, inflation_df['Tasa de Inflación'], color='blue')

# Configurar los títulos y etiquetas
plt.title('Tasa de Inflación en México (1996 - 2024)', fontsize=16)
plt.xlabel('Fecha', fontsize=14)
plt.ylabel('Tasa de Inflación (%)', fontsize=14)

# Mostrar la gráfica
plt.grid(True)
plt.show()

```

```python


import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Verificar si el índice ya es un DatetimeIndex
print(type(inflation_df.index))

# Configurar el tamaño de la gráfica
plt.figure(figsize=(10, 6))

# Graficar la tasa de inflación
plt.plot(inflation_df.index, inflation_df['Tasa de Inflación'], color='blue')

# Configurar los títulos y etiquetas
plt.title('Tasa de Inflación en México (1996 - 2024)', fontsize=16)
plt.xlabel('Fecha', fontsize=14)
plt.ylabel('Tasa de Inflación (%)', fontsize=14)

# Ajustar los ticks en el eje X para mostrar más meses
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=24))  # Mostrar cada mes
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))  # Formato de los ticks

# Rotar las etiquetas para mejor visibilidad
plt.gcf().autofmt_xdate()

# Mostrar la gráfica
plt.grid(True)
plt.show()


```

```python
from pandas.plotting import lag_plot
lag_plot (inflation_df, lag=20)
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Seleccionar la columna 'Tasa de Inflación' del DataFrame
inflation_series = inflation_df['Tasa de Inflación']

# Graficar la función de autocorrelación (ACF)
plot_acf(inflation_series, lags=24)
plt.show()
```

```python

from statsmodels.graphics.tsaplots import month_plot

# Asegúrate de que el índice sea una PeriodIndex con frecuencia mensual
inflation_df.index = pd.PeriodIndex(inflation_df.index, freq='M')

# Verificar las primeras filas para asegurarse de que los datos están correctos
print(inflation_df.head())
```
